Q1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
import selenium   
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

rank = []
name = []
udate = []
upload =[]
views = []
cell_value=[]

# Initialize WebDriver (for example, Chrome)
driver = webdriver.Chrome()

# Navigate to the web page containing the table
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos#Top_videos')

# 2 second delay
time.sleep(2)

#variable defiend and initialized
count=1

# Locate the table element
table = driver.find_element(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]//tbody')
rows = table.find_elements(By.TAG_NAME,'tr')

for row in rows:
    r=1
    cells = row.find_elements(By.TAG_NAME,'td')
    for cell in cells:
        
        cell_value = cell.text  # Extract text content of the cell
        
        if r == 5:
            rank.append(count)
            count=count+1
        if r == 1:
            name.append(cell_value)
        if r == 2:
            upload.append(cell_value)
        if r == 3:
            views.append(cell_value)
        if r== 4:
            udate.append(cell_value)
       
        r= r+1
        
data = {"Rank":rank,"Video Name ":name,"Artist":upload,"Views":views,"Date":udate}

df = pd.DataFrame(data)

#Close the browser
driver.quit()
df
```

Q2. Scrape the details team India‚Äôs international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:
A) Series
B) Place
C) Date
D) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
import selenium
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

all_url = []
tourna = []
match = []
venue = []
tdate = []
ttime = []

# Initialize WebDriver (for example, Chrome)
driver = webdriver.Chrome()

# Navigate to the web page containing the table
driver.get('https://www.bcci.tv/')

# 2 second delay
time.sleep(2)

# Locate the table element
search = driver.find_element(By.XPATH,'/html/body/header/div[3]/div[2]/ul/div[1]/a[2]')
search.click()
time.sleep(3)

fetchurl = driver.find_elements(By.XPATH,'//a[@class="match-btn match-card-btn ng-scope"]')
for f in fetchurl:
    url= f.get_attribute('href')
    all_url.append(url)
    
#print(len(all_url))

time.sleep(3)

for i in range(0,len(all_url)):
    driver.get(all_url[i])
    #time.sleep(10)
    WebDriverWait(driver, 10).until(
        EC.visibility_of_all_elements_located((By.XPATH,'//div[@class="matchDate ng-binding ng-scope"]')))

    fetch_date = driver.find_element(By.XPATH,'//div[@class="matchDate ng-binding ng-scope"]').text
    tdate.append(fetch_date)
    
    fetch_time = driver.find_element(By.XPATH,'//div[@class="matchTm ng-binding"]').text
    ttime.append(fetch_time)
      
    fetch_tour = driver.find_element(By.XPATH,'//li[@class="ng-scope"][1]/span[2]').text
    tourna.append(fetch_tour)
    
    fetch_match = driver.find_element(By.XPATH,'//li[@class="ng-scope"][2]/span[2]').text
    match.append(fetch_match)
    
    fetch_venue = driver.find_element(By.XPATH,'//li[@class="ng-scope"][3]/span[2]').text
    venue.append(fetch_venue)

df=pd.DataFrame({"DATE":tdate,
                 "TIME":ttime,
                 "TOURNAMENT":tourna,
                 "MATCH":match,
                 "PLACE":venue
                })
df
```

Q4. Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used

```python
import selenium
import requests
import re
from bs4 import BeautifulSoup
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

all_url = []
title = []
scrub_title=[]
description = []
contributors = []
scrub_contri = []
language =[]
lan1 = []

# Initialize WebDriver (for example, Chrome)
driver = webdriver.Chrome()

# Navigate to the web page containing the table
driver.get('https://github.com/')

# 3 second delay
time.sleep(5)

search = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button')
search.click()

time.sleep(5)

trending = driver.find_element(By.XPATH,'//ul[@aria-labelledby="open-source-repositories-heading"]/li[2]/a')
trending.click()
time.sleep(5)

url = driver.find_elements(By.XPATH,'//a[@class="Link"]')
for u in url:
    all_url.append(u.get_attribute('href'))

# All url are stored in the defined list.
    
pat="\n"    
for i in range (0,len(all_url)):
    response = requests.get(all_url[i])
    if response.status_code == 200:
        time.sleep(12)
        html_content= response.text
        soup = BeautifulSoup(html_content,'html.parser')
        try:
            temp_title = soup.find('strong',class_='mr-2 flex-self-stretch').text
            title.append(re.sub(pat,"",temp_title))
            
        except Exception as e:
            print("Title Element not found",e)
            title.append("-")
        try:
            temp_desc = soup.find('p',class_='f4 my-3')
            description.append(re.sub(pat,"",temp_desc.text))
            
        except Exception as e:
            print("Title Description Element not found", e)
            description.append("-")
            
        try:

            temp_contri = soup.find_all('span',class_='Counter ml-1')
            if len(temp_contri) > 1:
                contributors.append(temp_contri[1].text)
            else:
                contributors.append(temp_contri[0].text)
                       
        except Exception as e:
            print("contributor element not found", e)
            contributors.append("-")  
            
        try:
            temp_lang = soup.find_all('li',class_='d-inline')
            for li in temp_lang:
                lan1.append(re.sub(pat,"",li.text))
            
            language.append(lan1)
            
        except Exception as e:
            print("Language element not found",e)
            language.append("-")
    
    else:
        print("failed")

print(f'Title {len(title)} Description {len(description)} Contributors {len(contributors)} Language {len(language)}')        

data = {
        "Repository title":title,
        "Repository description":description,
        "Repository Contributors":contributors,
        "Language used":language
       }

df = pd.DataFrame(data)        
df 
```

Q6. Scrape the details of Highest selling novels.
A) Book name
B) Author name
C) Volumes sold
D) Publisher
E) Genre
Url - https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare

```python
import selenium
import requests
import re
from bs4 import BeautifulSoup
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

rank = []
book_name = []
author_name = []
volume_sold=[]
publisher = []
genre = []
cell_data = []

url ='https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare'

response = requests.get(url)
if response.status_code == 200:
    time.sleep(10)
    html_content= response.text
    soup = BeautifulSoup(html_content,'html.parser')
    
    table = soup.find('table',class_='in-article sortable')
    rows = table.find_all('tr')
    
    for row in rows:
        cells = row.find_all('td',class_="left")      
        r = 1
        for cell in cells:
            if r == 1:
                rank.append(cell.text)
            if r == 2:
                book_name.append(cell.text)
            if r == 3:
                author_name.append(cell.text)
            if r == 4:
                volume_sold.append(cell.text)
            if r == 5:
                publisher.append(cell.text)
            if r == 6:
                genre.append(cell.text)
            r= r+1    
                
            #cell_data.append(cell.text)           
               
        
                 
# print(len(book_name),len(author_name),len(volume_sold),len(publisher),len(genre))
                
data = {"Book Name":book_name,"Author Name": author_name,"Volume Sold":volume_sold,"Publisher":publisher,"Genre":genre}       
df = pd.DataFrame(data)        
df 
```
