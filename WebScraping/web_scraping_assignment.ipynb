#  1) Write a python program to display all the header tags from wikipedia.org and make data frame. linke : https://en.wikipedia.org/wiki/Main_Page" is used

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

def ws_heading(url):
    page = requests.get(url)
    if page.status_code == 200:
        soup = BeautifulSoup(page.content)
        for item in soup.find_all('span',class_="mw-headline"):
            heading.append(item.text)
        
        return heading
    else:
        print(f"please check the 'URL'")  
                   

url = "https://en.wikipedia.org/wiki/Main_Page"
heading = []
df = pd.DataFrame({'HEADLINES':ws_heading(url)})
df
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

def formal_presidents_name(url):
    page = requests.get(url)
    if page.status_code == 200:
        soup = BeautifulSoup(page.content)
        for item in soup.find_all('h3'):
            pname.append(item.text)
        
        return pname
    else:
        print(f"please check the 'URL'")  
        
def formal_presidents_Toffice(url):
    page = requests.get(url)
    if page.status_code == 200:
        soup = BeautifulSoup(page.content)
        for item in soup.find_all('h5'):
            toffice.append(item.text)
        
        return toffice
    else:
        print(f"please check the 'URL'")
        
                   

url = "https://presidentofindia.nic.in/former-presidents"
pname = []
toffice = []
df = pd.DataFrame({'Presedent Name':formal_presidents_name(url),'Term Of Office':formal_presidents_Toffice(url)})
df

```

Write a python program to scrape mentioned details from dineout.co.in and make data frame-
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

def restaurantname(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content)
    for item in soup.find_all('a',class_='restnt-name ellipsis'):
        rname.append(item.text)    
    return rname

def cuiname(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content)
    for item in soup.find_all('span',class_='double-line-ellipsis'):
        cuisine.append(item.text)    
    return cuisine

def restloc(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content)
    for item in soup.find_all('div',class_='restnt-loc ellipsis'):
        rloc.append(item.text)    
    return rloc

def restimage(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content)
    for item in soup.find_all('img',class_='no-img'):
        rimage.append(item['data-src'])    
    return rimage

    

rname = []
cuisine = [] 
rloc = []
rimage = []
url = "https://www.dineout.co.in/delhi-restaurants/welcome-back"
df = pd.DataFrame({'Resturant Name ': restaurantname(url),'Cuisine': cuiname(url), 'Location' : restloc(url), 'Image' : restimage(url)})
df
```
